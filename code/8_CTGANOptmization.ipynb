```python
import pandas as pd
from ctgan import CTGAN
from collections import Counter
import matplotlib.pyplot as plt
from table_evaluator import TableEvaluator
```

```python
cleaned_data = pd.read_csv('/content/drive/MyDrive/cleaned_data.csv')
cleaned_data.head()
```

```python
cleaned_data.shape
```

```python
class_counts = Counter(cleaned_data['Reason of Failure'])
class_counts
```

```python
labels = [f"Class {cls}" for cls in class_counts.keys()]
sizes = class_counts.values()

plt.figure(figsize=(8, 8))

# Custom function to display percentage only for Class 0
def autopct_func(pct):
    return f'{pct:.1f}%' if pct >= 96 else ''  # Show only if percentage is 96% or more

# Create pie chart
wedges, texts, autotexts = plt.pie(
    sizes, labels=None, startangle=120, colors=plt.cm.Paired.colors,
    autopct=autopct_func,  # Apply custom function
    textprops={'fontsize': 12}  # Adjust font size
)

# Legend
plt.legend(wedges, labels, title="Classes", loc="center left", bbox_to_anchor=(1, 0.5))

plt.title("Class Distribution After Data Augmentation")
plt.show()
```

```python
class_counts = Counter(cleaned_data['Reason of Failure'])
class_counts
```

```python
max_class_size = max(class_counts.values())
max_class_size
```

```python
ctgan_augmented_data = []
```

```python
ctgan = CTGAN(generator_lr = 0.001,
              discriminator_lr = 0.001,
              batch_size = 200,
              epochs = 500,
              verbose = True
              )
```

```python
for class_label in class_counts.keys():
    class_data = cleaned_data[cleaned_data['Reason of Failure'] == class_label]
    ctgan.fit(class_data)
    synthetic_class_data = ctgan.sample(max_class_size - len(class_data))
    ctgan_augmented_data.append(synthetic_class_data)
```

```python
balanced_data = pd.concat([cleaned_data] + ctgan_augmented_data)
balanced_data.head()
```
